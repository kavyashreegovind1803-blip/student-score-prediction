#DAY 5- DATA CLEANING

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Student": ["Asha", "Rahul", "Priya", "Ravi", "Sneha", "Kiran", "Anjali", "Vijay", "Deepa", "Arjun"],
    "Hours_Studied": [2, 5, 3, 8, 6, 7, 4, 9, 5, 10],
    "Attendance": [85, 90, 80, 95, 88, 92, 84, 98, 89, 97],
    "Assignment_Score": [70, 82, 65, 94, 80, 86, 75, 96, 81, 98],
    "Final_Score": [65, 80, 60, 95, 82, 88, 72, 97, 83, 99]
}

df = pd.DataFrame(data)
print(df)
```

```python
print("\nFirst Five Records")
print(df.head())

print("\nLast Five Records")
print(df.tail())

print("\nDataset Information")
print(df.info())

print("\nStatistical Summary")
print(df.describe())
```

```python
print("\nStudents who studied more than 5 hours")
print(df[df["Hours_Studied"] > 5])
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Hours_Studied"], df["Final_Score"], marker="o")
plt.title("Hours Studied vs Final Score")
plt.xlabel("Hours Studied")
plt.ylabel("Final Score")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Student"], df["Final_Score"])
plt.title("Student Final Scores")
plt.xlabel("Students")
plt.ylabel("Final Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Hours_Studied"], df["Final_Score"])
plt.title("Hours Studied vs Final Score")
plt.xlabel("Hours Studied")
plt.ylabel("Final Score")
plt.show()
```

```python
print("Average Score:", df["Final_Score"].mean())
print("Highest Score:", df["Final_Score"].max())
print("Lowest Score:", df["Final_Score"].min())

print("\nTop Performer")
print(df.loc[df["Final_Score"].idxmax()])
```
